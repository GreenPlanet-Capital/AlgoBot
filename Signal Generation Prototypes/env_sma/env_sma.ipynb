# Simple Envelope

**Formula:**
* Upper envelope = SMANx X [1 + D%]
* Lower envelope = SMANx X [1 - D%]

**Usage:**
* A moving average smooths out price fluctuations and allows us to see the broader pattern of the market. A moving average that slopes upward confirms that prices have been trending upwards. * Conversely, if a MA line slopes downward, it indicates a downtrend. With our Envelopes indicator, we can also look at the direction of our bands to inform us about the trend. If our bands are sloping upward, then it confirms an uptrend. If our bands are sloping downward, it confirms a downtrend.
* As a simple trend-following signal, we can look for those times when the current price crosses above a moving average line. This may be a signal for a breakout into a new upward trend. Likewise, a downward crossover of the price through the moving average may signal a new downtrend. We can use the Envelopes indicator in a similar manner. With our Envelopes, they display MA lines that have been shifted up and down.

**Notes:** 
* We can use EMA instead of SMA

**Pseudo-Code Definition:** envelope_scan(timeseries)

```python
def env_sma(dataframe_input, lookback_period, d_percentage = 3):
    
    df = dataframe_input
    n = lookback_period
    
    df['TYP PRICE'] =  (df['CLOSE'] +  df['LOW'] +  df['HIGH'] +  df['OPEN'])/4
    
    df_indicators = pd.DataFrame()
    
    df_indicators['DATE'] = df['DATE']
    
    temp_list = [None for i in range(len(df))]
    indic_columnhead = 'ENVSMA ' + str(lookback_period) + ' ' + str(d_percentage)
    df_indicators[indic_columnhead] = temp_list
    
    sma = [None for i in range(n)]
    env_up = [None for i in range(n)]
    env_down = [None for i in range(n)]
    
    initial_start_ctr = 0
    initial_end_ctr = n
    
    for i in range(len(df) - n):
        
        lookback_list = list(df['TYP PRICE'].iloc[initial_start_ctr : initial_end_ctr])
        sigma = d_percentage/100
        
        sum_var = sum(lookback_list)
        sma_val = sum_var/n
        
        env_up.append(sma_val * (1 + sigma))
        env_down.append(sma_val * (1 - sigma))
        sma.append(sma_val)
               
        initial_start_ctr += 1
        initial_end_ctr += 1

    df_indicators[indic_columnhead] = sma
    df_indicators['ENVUP' + indic_columnhead] =  env_up
    df_indicators['ENVDOWN' + indic_columnhead] = env_down
    
    return df_indicators
    
```

```python
def signal_generation(df, indic_df, lookback, d_percentage, sensitivity = 1):
    df_internal = pd.DataFrame()
    n = lookback
    m = d_percentage
    
    df_internal['TYP PRICE'] =  (df['CLOSE'] +  df['LOW'] +  df['HIGH'] +  df['OPEN'])/4
    df_internal['UPBOUND'] = indic_df['ENVUPENVSMA ' + str(n) + ' ' + str(m)]
    df_internal['DOWNBOUND'] = indic_df['ENVDOWNENVSMA ' + str(n) + ' ' + str(m)]
    df_internal['MIDLINE'] = indic_df['ENVSMA ' + str(n) + ' ' + str(m)]
    
    df_internal['UPBOUND_DIST'] = abs(df_internal['TYP PRICE'] - df_internal['UPBOUND'])
    df_internal['DOWNBOUND_DIST'] = abs(df_internal['DOWNBOUND'] - df_internal['TYP PRICE'])
    df_internal['UPBREAKOUT_SIGNAL'] = df_internal['TYP PRICE'] > df_internal['UPBOUND']
    df_internal['DOWNBREAKOUT_SIGNAL'] = df_internal['TYP PRICE'] < df_internal['DOWNBOUND']
    
    upbound_dist_max = df_internal['UPBOUND_DIST'].max()
    downbound_dist_max = df_internal['DOWNBOUND_DIST'].max()
    
    price_list = list(df_internal['TYP PRICE'])
    upbound_dist_list = list(df_internal['UPBOUND_DIST'])
    downbound_dist_list = list(df_internal['DOWNBOUND_DIST'])
    upbreakout_list = list(df_internal['UPBREAKOUT_SIGNAL'])
    downbreakout_list = list(df_internal['DOWNBREAKOUT_SIGNAL'])
    mid_list = list(df_internal['MIDLINE'])
    interest_list = []
    
    for i in range(n + 1,len(indic_df)):
        long_interest = 0
        short_interest = 0
        if (upbreakout_list[i]):
            long_interest = upbound_dist_max + upbound_dist_list[i]
            short_interest = 0
        elif (downbreakout_list[i]):
            short_interest = downbound_dist_max + downbound_dist_list[i]
            long_interest = 0
        else:
            short_interest = upbound_dist_max - upbound_dist_list[i]
            long_interest = downbound_dist_max - downbound_dist_list[i]
            if (price_list[i] <= mid_list[i]):
                short_interest = short_interest/2
            elif (price_list[i] > mid_list[i]):
                long_interest = long_interest/2
        non_linear_projection = long_interest - short_interest
        interest_list.append(non_linear_projection)
        
    #scaling
    a = min(interest_list)
    b = max(interest_list)
    b_dash = 100
    a_dash = -100
    scaled_interest_list_out = [None for i in range(n + 1)]
    scaled_interest_list = []
    for i in interest_list:
        frac = (i - a)/(b - a)
        val1 = frac*(b_dash - a_dash)
        scaled_val = val1 + a_dash
        scaled_interest_list.append(scaled_val)
        scaled_interest_list_out.append(scaled_val)
        
    df_out = pd.DataFrame()
    df_out['DATE'] = indic_df['DATE']
    df_out['ENVSMA SIGNAL ' + str(n) + ' ' + str(m)] = scaled_interest_list_out
    
    #signum truth table construction
    indic_mean = df_out['ENVSMA SIGNAL ' + str(n) + ' ' + str(m)].mean()
    absolute_mean = 0
    indic_std = df_out['ENVSMA SIGNAL ' + str(n) + ' ' + str(m)].std()
    absolute_std = 85
    
    df_internal['ENVSMA SIGNUM BUY ' + str(n) + ' ' + str(m)] = df_out['ENVSMA SIGNAL ' + str(n) + ' ' + str(m)] >  (indic_mean + indic_std * sensitivity)
    df_internal['ENVSMA SIGNUM SELL ' + str(n) + ' ' + str(m)] = df_out['ENVSMA SIGNAL ' + str(n) + ' ' + str(m)] <=  (indic_mean - indic_std * sensitivity)
    df_internal['ABSOLUTE ENVSMA SIGNUM BUY ' + str(n) + ' ' + str(m)] = df_out['ENVSMA SIGNAL ' + str(n) + ' ' + str(m)] >  (absolute_mean + absolute_std * sensitivity)
    df_internal['ABSOLUTE ENVSMA SIGNUM SELL ' + str(n) + ' ' + str(m)] = df_out['ENVSMA SIGNAL ' + str(n)+ ' ' + str(m)] <=  (absolute_mean - absolute_std * sensitivity)
   
    #indicator signum
    long = list(df_internal['ENVSMA SIGNUM BUY ' + str(n) + ' ' + str(m)])
    short = list(df_internal['ENVSMA SIGNUM SELL ' + str(n) + ' ' + str(m)])
    
    indic_out = [] 
    for i in range(len(long)):
        append_val = 0
        if (long[i] == True and short[i] == False):
            append_val = 100
        elif (long[i] == False and short[i] == True):
            append_val = -100
        else:
            append_val = 0 
        indic_out.append(append_val)
        
    df_out['ENVSMA SIGNUM ' + str(n) + ' ' + str(m)] = indic_out
    
    #absolute signum
    abs_long = list(df_internal['ABSOLUTE ENVSMA SIGNUM BUY ' + str(n) + ' ' + str(m)])
    abs_short = list(df_internal['ABSOLUTE ENVSMA SIGNUM SELL ' + str(n) + ' ' + str(m)])
    
    abs_out = [] 
    for i in range(len(long)):
        append_val = 0
        if (abs_long[i] == True and abs_short[i] == False):
            append_val = 100
        elif (abs_long[i] == False and abs_short[i] == True):
            append_val = -100
        else:
            append_val = 0 
        abs_out.append(append_val)
    
    df_out['ABSOLUTE ENVSMA SIGNUM ' + str(n) + ' ' + str(m)] = abs_out
    
    return df_out
```

```python
signal_df = signal_generation(df = df, indic_df = indic_df1, lookback = lookback1, d_percentage = d_percentage, sensitivity = 1)
print(signal_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.text

def test_plot_signal(data, indicator_frame, signal_frame, lookback, d_percentage):
    # Load up the dataframe    
    df1 = data
    df2 = indicator_frame
    df3 = signal_frame
    col1 = 'ENVSMA' + ' ' + str(lookback1) + ' ' + str(d_percentage)
    sig_col = 'ENVSMA SIGNAL ' + str(lookback) + ' ' + str(d_percentage)
    signum_col = 'ENVSMA SIGNUM ' + str(lookback) + ' ' + str(d_percentage)
    absolute_signum_col = 'ABSOLUTE ENVSMA SIGNUM ' + str(lookback) + ' ' + str(d_percentage)
    
    # Set up objects for plotting an subplots
    fig, axes = plt.subplots(5, 1, figsize=(20, 20), sharex=True)
    fig.suptitle('Indicator Comparison')
    sns.set_style("darkgrid")

    #Plotting Prices
    sns.lineplot(ax=axes[0], x = df1['DATE'], y = df1['LOW'], data = df1, color = "cyan")
    sns.lineplot(ax=axes[0], x = df1['DATE'], y = df1['HIGH'], data = df1, color = "blue")
    axes[0].set_title("Price")

    # Plotting the Indicator
    sns.lineplot(ax=axes[1],x = df2['DATE'], y = df2[col1], data = df2, color = "orange")
    sns.lineplot(ax=axes[1],x = df2['DATE'], y = df2['ENVUP'+col1], data = df2, color = "blue")
    sns.lineplot(ax=axes[1],x = df2['DATE'], y = df2['ENVDOWN'+col1], data = df2, color = "blue")
    sns.lineplot(ax=axes[1], x = df1['DATE'], y = df1['OPEN'], data = df1, color = "black")
    axes[1].set_title("Indicator")
    
    sns.lineplot(ax=axes[2],x = df3['DATE'], y = df3[sig_col], data = df3, color = "red")
    axes[2].set_title("Signal")
    
    sns.lineplot(ax=axes[3],x = df3['DATE'], y = df3[signum_col], data = df3, color = "darkgreen")
    axes[3].set_title("Signum")
    
    sns.lineplot(ax=axes[4],x = df3['DATE'], y = df3[absolute_signum_col], data = df3, color = "green")
    axes[4].set_title("Absolute Signum")
```
